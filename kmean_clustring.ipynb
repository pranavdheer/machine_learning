```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
from sklearn.cluster import KMeans
import pandas as pd
```

```python
d=pd.read_csv('./machine_learnings/toy.csv')
```

```python
print d.shape
print d.columns
```

```python
x=d.values
print x[1,2]
plt.figure(0)    #graph to label
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
#final output
plt.figure(0)
for ix in range (100):
    if x[ix,2]==1:         #expected output
      
        plt.scatter(x[ix,0],x[ix,1],color='red')
    else:
        plt.scatter(x[ix,0],x[ix,1],color='green')
plt.show()
x_train=x[:,:2]
print x_train.shape
```

```python
km = KMeans(n_clusters=3)
p= km.fit(x_train)
print p
pred = km.predict(x_train)
print pred
```

```python
plt.figure(0)
for ix in range (pred.shape[0]):
        if pred[ix] == 0:
            plt.scatter(x_train[ix, 0], x_train[ix, 1], color='red')
        elif pred[ix] == 1:
            plt.scatter(x_train[ix, 0], x_train[ix, 1], color='green')
        else:
            plt.scatter(x_train[ix, 0], x_train[ix, 1], color='blue')

plt.show()
```

```python
h=np.asarray([[-5,2],[0,4]])
km = KMeans(n_clusters=2,init=h)
p= km.fit(x_train)
pred = km.predict(x_train)
print pred
```

```python
plt.figure(0)
for ix in range (pred.shape[0]):
        if pred[ix] == 0:
            plt.scatter(x_train[ix, 0], x_train[ix, 1], color='red')
        elif pred[ix] == 1:
            plt.scatter(x_train[ix, 0], x_train[ix, 1], color='green')
        else:
            plt.scatter(x_train[ix, 0], x_train[ix, 1], color='blue')

plt.show()
```
